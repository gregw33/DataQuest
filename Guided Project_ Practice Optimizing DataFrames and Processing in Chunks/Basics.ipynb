```python
# Prep for which are string cols
dtypes = df_initial.dtypes
str_cols = dtypes[dtypes == 'object'].index.to_list()
print(str_cols)
# print(df_initial.loc[:,str_cols])
```

```python
# prep for which float cols
float_cols = dtypes[dtypes == 'float64'].index.to_list()
print(float_cols)
```

```python
chunk_iter = pd.read_csv('loans_2007.csv', chunksize=3000)
str_uniques = {col: [] for col in str_cols}
float_missing = {col: 0 for col in float_cols}
float_badint = {col: 0 for col in float_cols}
mems = 0 # in Mb
for chunk in chunk_iter:
    mems += chunk.memory_usage(deep=True).sum() / 2**20
    for col in str_cols:
        val_counts = chunk[col].value_counts()
        uniques = (val_counts==1).sum()
        str_uniques[col].append(uniques)
    for col in float_cols:
        nans = chunk[col].isna()
        float_missing[col] += nans.sum()
        badints = (chunk[col] // 1 != 0) & ~nans
        float_badint[col] += badints.sum()
str_sums = {col: sum(vals) for col, vals in str_uniques.items()}
```

```python
print(mems, 'MB used across the entire dataset')
display(str_sums)
display(float_missing)
display(float_badint)
```

```python
# Cols which can be converted to categories (probably)
category_potentials = []
for col, val in str_sums.items():
    if val < 1000:
        category_potentials.append(col)
print(category_potentials)
```
